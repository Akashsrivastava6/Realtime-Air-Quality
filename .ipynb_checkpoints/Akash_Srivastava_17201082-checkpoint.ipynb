```python
#importing all the packages needed for data collection
import urllib.request #used to get the data from web.
import csv #used to write the data into a file
import pandas as pd # used for working with the data from the file as data frame
import os
```

By using the above key the data is collected. For the purpose of data collection, Collecting_data function is written.

```python
#following function is used for collecting the data from the web
def Collecting_data(Api_link):
    path=urllib.request.urlopen(Api_link) # specifying the path to get the data
    data=path.read().decode()
    data_list=data.split("\n") # spliting each row by '\n'.
    return data_list 


# following function is used to write data into a file.
def writing_data(Api_link,file_name):
    print("Data Collection Started...\n")
    dat_list=Collecting_data(Api_link) # collecting_data function is called for data collection
    print("Data collection completed. Now Saving this data into file.\n")
    file=open(file_name,"w") # csv file is oped to write data into it.
    wrt=csv.writer(file,delimiter=",") # comma is used as delimiter as the file format chosen is csv.
    for a in csv.reader(dat_list): 
        wrt.writerow(a) # Each row is added into the file
    file.close()
```

```python
# below are the functions that returns the color of the graph based on the values of NO2, SO2 and PM10

#this function return color for NO2 values
def get_color_NO2(chk):
    color_NO2=[]
    for a in chk:
        if a < 50:
            color_NO2.append('green')
        elif a>=50 and a<60:
            color_NO2.append('olive')
        
        elif a>=60 and a<75:
            color_NO2.append('orange')
        elif a>=75 and a<100:
            color_NO2.append('red')
        else:
            color_NO2.append('firebrick')

    return color_NO2


# this function returns color for SO2 values
def get_color_SO2(chk):
    color_SO2=[]
    for a in chk:
        if a < 3:
            color_SO2.append('green')

        elif a>=3 and a<5:
            color_SO2.append('olive')
        elif a>=5 and a<7:
            color_SO2.append('orange')
        elif a>=7 and a<9:
            color_SO2.append('red')
        else:
            color_SO2.append('firebrick')
    return color_SO2

# this function returns color for SO2 values
def get_color_PM10(chk):        
    color_PM10=[]
    for a in chk:
        if a < 200:
            color_PM10.append('green')

        elif a>=201 and a<251:
            color_PM10.append('olive')
        elif a>=251 and a<351:
            color_PM10.append('orange')
        elif a>=351 and a<500:
            color_PM10.append('red')
        else:
            color_PM10.append('firebrick')
    return color_PM10

# this function returns legend for the charts
def get_legends():
    grn=mpatches.Patch(color='green',label='Good')
    ylw=mpatches.Patch(color='olive',label='Moderate')
    org=mpatches.Patch(color='orange',label='Unhealthy for sensitive group')
    rd=mpatches.Patch(color='red',label='Unhealthy')
    fb=mpatches.Patch(color='firebrick',label='Very Unhealthy')
    
    return [grn,ylw,org,rd,fb]
```

```python
# grouping realtime data by state

data_realtime_state=data_realtime.groupby(data_realtime['state'])# grouping realtime data by state

data_realtime_state=data_realtime_state.describe() # using describe() to get mean
data_realtime_state
```

```python
#importing packages required for generating map
from mpl_toolkits.basemap import Basemap 
```

```python
# ploting NO2 levels at various states in India
plt.subplots(figsize=(20, 15))
map = Basemap(width=1200000,height=900000,projection='aea',resolution='l',
                    llcrnrlon=67,llcrnrlat=5,urcrnrlon=99,urcrnrlat=37,lat_0=28,lon_0=77) # creating map 


map.drawcountries(color='grey') # used for drawing countries boundries
map.drawcoastlines() # used for drawing coastline

# taking out longitude of each state from the data_realtime_state dataframe.
lg=data_realtime_state['longitude'].iloc[:,data_realtime_state['longitude'].columns.get_level_values(0)=='mean'].values.flatten()
# taking out latitude of each state from the data_realtime_state dataframe.
lt=data_realtime_state['latitude'].iloc[:,data_realtime_state['latitude'].columns.get_level_values(0)=='mean'].values.flatten()
# taking out NO2 value of each state from the data_realtime_state dataframe.
pt=data_realtime_state['NO2_avg'].iloc[:,data_realtime_state['NO2_avg'].columns.get_level_values(0)=='mean'].values.flatten()
state_name=data_realtime_state.index #this variable is used for xticks

# plotting points on map for each state
x, y = map(lg, lt)
cs=plt.scatter(x, y,s=150,color=get_color_NO2(pt), alpha=0.7)

# Adding text with points on map
for state_names, xpt, ypt in zip(state_name, x, y):
    plt.text(xpt+60000, ypt+30000, state_names, fontsize=10, fontweight='bold')
    
# Adding legend to chart
plt.legend(handles=get_legends())

# Adding title to chart
plt.title('Current NO2 level at various states in India',fontsize=20)
plt.show()
```
